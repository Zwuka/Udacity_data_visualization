```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.histplot(data = df,x='duration_sec',color = color);

plt.subplot(1,2,2)
sns.histplot(data = df,x='duration_sec',color = color);
plt.xlim(0,2500);
```

In cause that we have long tale let's make log transform

```python
bins = 10** np.arange(1.9,np.log10(df.duration_sec).max() +0.025,0.025)
plt.hist(data = df,x='duration_sec', bins = bins);
plt.xscale('log');
ticks = [50,100,250,500,1000,3000,10000,25000,85000]
labels = [str(x) if x<1000 else '{0:.0f}K'.format(x/1000) for x in ticks]
plt.xticks(ticks,labels);
```

 When plotted on a log scale, the time of ride looks roughly normal, with a long tail on the right side.

Let's look deeper in birth year of our subscribers

```python
plt.figure(figsize=(20,6))
plt.subplot(1,3,1)
df.member_birth_year.hist()

plt.subplot(1,3,2)
df[df.member_birth_year <1960].member_birth_year.hist();

plt.subplot(1,3,3)
df[df.member_birth_year <1930].member_birth_year.hist();
```

I think there are errors in the dates of birth before 1930, but I'm not entirely sure about the years in the range from 1920 to 1930 (maybe there are some strong old people riding bicycles, or maybe their children use parent's accounts). But I am absolutely sure that the records with the years of birth in the range from 1880 to 1920 were not created correctly.
Let's replace incorrect years of birth with NaN values

```python
df.loc[df['member_birth_year'] <1921,['member_birth_year']] = np.NaN
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
df.member_birth_year.hist()

plt.subplot(1,2,2)
sns.histplot(data = df, x = 'member_birth_year',bins = 35);
plt.xticks(np.arange(1930,2005,4),rotation = 90);

```

```python
df.member_birth_year.describe()
```

As we see in plot, most of our subscribers are born around 1984-1994 years

Let's make deeper look at location where bike was rented:

```python
plt.figure(figsize=(21,6))
plt.subplot(1,3,1)
sns.histplot(data = df, x = 'start_station_id');

plt.subplot(1,3,2)
sns.histplot(data = df, x = 'start_station_longitude');

plt.subplot(1,3,3)
sns.histplot(data = df, x = 'start_station_latitude');
```

```python
b = df[df.start_station_longitude < -122.2]['start_station_id'].nunique() #how much stations we have in longitude lower then -122.2
c = df['start_station_id'].nunique() #how much station we have at all
print ('We have {} station that longitude lower then -122.2, there {} station at all. So {}% of stations stays in this area'.format(b,c,"{0:.1f}".format(b/c*100)))
```

```python
df.start_station_longitude.describe()
```

I think we have same distribution in end station's but let's check for sure

```python
plt.figure(figsize=(21,6))
plt.subplot(1,3,1)
sns.histplot(data = df, x = 'end_station_id');

plt.subplot(1,3,2)
sns.histplot(data = df, x = 'end_station_longitude');

plt.subplot(1,3,3)
sns.histplot(data = df, x = 'end_station_latitude');
```

So as I expected station's are grouped in same areas. Let's make new station's id in dependence of their longitude.

```python
list_end_stations = list(df.end_station_longitude.unique())
list_end_stations.sort()
list_start_stations = list(df.start_station_longitude.unique())
list_start_stations.sort()
```

```python
len(list_start_stations), len(list_end_stations), list_start_stations == list_end_stations
```

Everything works like i expected. So let's rework id using station longitude. By doing this work we also remove Nan data in station id.

```python
df['start_station_id_by_long'] = df.start_station_longitude.apply(lambda x: list_start_stations.index(x)+1)
df['end_station_id_by_long'] = df.end_station_longitude.apply(lambda x: list_start_stations.index(x)+1)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.histplot(df.start_station_id_by_long);

plt.subplot(1,3,2)
sns.histplot(df.end_station_id_by_long);

plt.subplot(1,3,3)
sns.histplot(df.end_station_id_by_long);
plt.xlim(120,150);
```

```python
sns.histplot(data = df,x = 'start_station_id_by_long');
```

Interesting thing here, a lot of rides end in stations 130-140.

Let's look at distribution ob rides between months. Also let's check is all rides done in one year or not.

```python
df['start_month'] =  df.start_time.dt.month
df['end_month'] = df.end_time.dt.month
df.start_month = df.start_month.astype('category')
df.end_month = df.end_month.astype('category')
df.start_month.value_counts(), df.end_month.value_counts()
```

```python
df.start_time.dt.year.value_counts()
```

Thus, all trips started in February 2019 and ended this month as well, with the exception of 16 trips that started in February 2019 and ended in March 2019.

Let's look on distribution of ride's made in every week.

```python
df['start_week_num'] = df.start_time.dt.isocalendar().week
df['start_day_num'] = df.start_time.dt.day
df['end_week_num'] = df.end_time.dt.isocalendar().week
df['end_day_num'] = df.end_time.dt.day
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.countplot(data = df, x = 'start_week_num',color = color);

plt.subplot(1,2,2)
sns.countplot(data = df, x = 'end_week_num',color = color);
```

```python
df.groupby(['start_week_num'])['start_day_num'].nunique()
```
